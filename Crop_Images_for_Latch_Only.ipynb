```


# Improting Image class from PIL module 
from PIL import Image 
  
# Opens a image in RGB mode 
im = Image.open(IMG_PATH) 
  
# Size of the image in pixels (size of orginal image) 
# (This is not mandatory) 
width, height = im.size 
  
# Setting the points for cropped image 
left = 1150
top = 500
right = 1374
bottom = 724
  
# Cropped image of above dimension 
# (It will not change orginal image) 
im1 = im.crop((left, top, right, bottom)) 
  
# Shows the image in image viewer 
display(im1)

im1.save('crop_Open0555.jpg', 'JPEG')
```

```
from PIL import Image
import os.path, sys

path = "/content/gdrive/My Drive/Colab_Files/CVFinal/Open/"
path2 = "/content/gdrive/My Drive/Colab_Files/CVFinal2/Open/"

dirs = os.listdir(path)

left = 1150
top = 500
right = 1374
bottom = 724

def crop():
    for item in dirs:
        fullpath = os.path.join(path,item)         #corrected
        if os.path.isfile(fullpath):
            im = Image.open(fullpath)
            f, e = os.path.splitext(fullpath)
            imCrop = im.crop((left, top, right, bottom)) #corrected
            imCrop.save(path2 + item + 'Cropped.png', "PNG", quality=95)

crop()
```
